```python
# importing all the required libraries

import zipfile
import os
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
# Specifying the path of the cotton weed dataset zip file
cotton_weed_path = '/content/drive/MyDrive/weed_identification/cottonweeds.zip'

with zipfile.ZipFile(cotton_weed_path, 'r') as zip_ref:
    # Extracting all the contents of the cotton weed data zip file to the current directory
    zip_ref.extractall()
```

```python
# Specifying the path of the deep weed dataset zip file
deep_weed_path = '/content/drive/MyDrive/weed_identification/deepweeds.zip'

with zipfile.ZipFile(deep_weed_path, 'r') as zip_ref:
    # Extracting all the contents of the deep weed data zip file to the current directory
    zip_ref.extractall()
```

```python
# Initializing empty lists to store image paths and labels
cottonweed_image_paths  = [] # List to store paths of images
cottonweed_image_labels = [] # List to store corresponding labels

# Iterate through all files and directories in the cottonweedID15 directory
for root, dirs, files in os.walk('/content/CottonWeedID15'):
    # Looping through each file in the current directory
    for file in sorted(files):
      # Getting the full file path
      cottonweed_file_path = os.path.join(root, file)
      # Appending the file path to the list of image paths
      cottonweed_image_paths.append(cottonweed_file_path)
      # Extracting the label from the file path and append it to the list of labels
      cottonweed_image_labels.append(cottonweed_file_path.split('/')[-2])
```

```python
# Create a DataFrame using the zip of cottonweed_image_paths and cottonweed_image_labels
# Each row in the DataFrame will contain an image path and its corresponding label

cottonweed_data = pd.DataFrame(list(zip(cottonweed_image_paths,cottonweed_image_labels)),
                               columns=['image_paths','image_labels'])
# Displaying first 5 rows of the data.
cottonweed_data.head()
```

```python
cottonweed_data.shape
```

There are 5187 images in the cottonweedID15 dataset.

```python
# Splitting the data into equal halves with equal number of class labels.
# and considering the first half of the dataset for our study.

cottonweed_data_split, _ = train_test_split(cottonweed_data, test_size=0.5, stratify=cottonweed_data['image_labels'],
                                            random_state=42)
```

```python
# Displaying the first 5 rows of the data.
cottonweed_data_split.head()
```

```python
cottonweed_data_split.shape
```

There are 2593 images in the first half of the cottonweedID15 data considered.

```python
# Reading labels file
deepweeds_data = pd.read_csv('/content/labels/labels.csv')
# Displaying the first 5 rows of the data.
deepweeds_data.head()
```

```python
# Renaming the columns of filenames and species as image_paths and image_labels respectively.
new_column_names = {'Filename': 'image_paths', 'Species': 'image_labels'}
deepweeds_data = deepweeds_data.rename(columns=new_column_names)
# Displaying the first 5 rows of the data.
deepweeds_data.head()
```

```python
# Dropping the label column which is insignificant
deepweeds_data.drop(['Label'],axis=1,inplace=True)
```

```python
# Appending the '/content/images/' to all image paths in deepweed data
deepweeds_data['image_paths'] = deepweeds_data['image_paths'].apply(lambda x: '/content/images/' + x)
# Displaying the first 5 rows of the data.
deepweeds_data.head()
```

```python
deepweeds_data.shape
```

There are 17509 images in the deepweed dataset.

```python
# Merging the two datasets [first half of the cottonweedID15 data + deepweed dataset] into one
merged_data = pd.concat([cottonweed_data_split,
                         deepweeds_data], axis=0,ignore_index=True)
# Displaying the first 5 rows of the data.
merged_data.head()
```

```python
merged_data.shape
```

The merged dataset has 20,102 images.

```python
# Checking the distribution of the class labels using pandas bar chart

plt.figure(figsize=(12,4))
label_counts = merged_data['image_labels'].value_counts()

# Plotting bar plot for distribution of class labels
ax = label_counts.plot(kind='bar')
plt.title('Distribution of class labels',color='red',size=12)
plt.xlabel('Labels',color='blue',size=12)
plt.ylabel('Counts',color='blue',size=12)

# Adding labels on the bars
for i, count in enumerate(label_counts):
    plt.text(i, count, str(count), ha='center', va='bottom',color='green')

plt.show()
```

**Observations:**<br>
* Class label - Negative has the highest number of images while SpurredAnoda has the least number of images.

```python
# Plotting few sample images from the dataset.

# Getting sample data
sample_image_data = merged_data.sample(n=20)

# Plotting the sample images
plt.figure(figsize=(15, 10))
for i, (image_path, image_label) in enumerate(zip(sample_image_data['image_paths'], sample_image_data['image_labels'])):
    plt.subplot(4, 5, i+1)
    image = Image.open(image_path)
    plt.imshow(image)
    plt.title(image_label)
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Checking the distribution of image heights and widths of all images.

# Initialize empty lists to store image heights and widths
heights = []
widths = []

# Get heights and widths of all images
for image_path in merged_data['image_paths']:
    image = Image.open(image_path)
    width, height = image.size
    widths.append(width)
    heights.append(height)

# Plot histogram of image widths
plt.figure(figsize=(15, 4))
plt.subplot(1,2,1)
plt.hist(widths, bins=20, color='green', edgecolor='red', alpha=0.8)
plt.title('Histogram of Image Widths',color='red')
plt.xlabel('Width',color='blue')
plt.ylabel('Frequency',color='blue')
plt.grid('--')

# Plot histogram of image heights
plt.subplot(1,2,2)
plt.hist(heights, bins=20, color='green', edgecolor='red', alpha=0.8)
plt.title('Histogram of Image Heights',color='red')
plt.xlabel('Height',color='blue')
plt.ylabel('Frequency',color='blue')
plt.grid('--')
plt.show()
```

**Observations:**<br>
* Most of image heights and widths approximately lie in the range of 200-500.

```python
# Summary statistics of the image data.

# Initialize lists to store summary statistics
intensity_means = []
intensity_stds = []

# Loop through each image path in the DataFrame
for image_path in merged_data['image_paths']:
    # Load the image
    image = Image.open(image_path)

    # Convert the image to grayscale
    gray_image = image.convert('L')

    # Convert the image to a NumPy array
    image_array = np.array(gray_image)

    # Calculate summary statistics of image intensities
    intensity_mean = np.mean(image_array)
    intensity_std = np.std(image_array)

    # Append the statistics to the lists
    intensity_means.append(intensity_mean)
    intensity_stds.append(intensity_std)

# Calculate overall summary statistics
overall_mean = np.mean(intensity_means)
overall_std = np.mean(intensity_stds)

print("Overall Mean Intensity:", overall_mean)
print("Overall Standard Deviation of Intensity:", overall_std)
```

The overall mean intensity and standard deviation of intensity derived from the images encapsulate statistical summaries of pixel intensity values across the entire image dataset.

**Overall Mean Intensity:** This figure, approximately 104.06, signifies the average intensity level spanning all pixels within the images. Put differently, it represents the collective brightness across the images. A higher mean intensity indicates a tendency towards brighter pixels on average, while a lower mean intensity suggests darker pixel values prevail.

**Overall Standard Deviation of Intensity:** At approximately 55.22, this metric reflects the dispersion or variability of pixel intensity values around the mean intensity. A higher standard deviation suggests a wider spread of pixel intensity values from the mean, indicating greater variability or contrast among the images. Conversely, a lower standard deviation indicates pixel intensity values clustering closer to the mean, implying less variability or contrast.

In essence, these statistical measures offer insights into the overall brightness and contrast characteristics exhibited by the images in your dataset.
